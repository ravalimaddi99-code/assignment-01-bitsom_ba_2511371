## Part 4: Semantic Embeddings with Cosine Similarity

**Objective:** Use sentence-transformer embeddings to represent sentences as vectors, compute cosine similarity, and visualize relationships using a heatmap.


```python
# Step 1: Install required library
!pip install sentence-transformers
```

```python
# Step 2: Import libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

## Step 3: Define Sentences (3 Topics)

```python
# Step 3: Create 10 sentences (3 topics: Cricket, Cooking, Cybersecurity)
sentences = [
    # Cricket
    "The bowler took three wickets in one over",
    "The batsman scored a century in the final match",
    "The team celebrated their victory after the match",

    # Cooking
    "I cooked a delicious biryani for dinner",
    "The recipe requires fresh vegetables and spices",
    "She baked a chocolate cake in the oven",

    # Cybersecurity
    "Strong passwords help protect your online accounts",
    "Cyber attacks can compromise sensitive data",
    "Encryption ensures secure communication over the internet",

    # Extra (to make 10)
    "Firewalls help in preventing unauthorized access"
]

print(f"Total sentences: {len(sentences)}")
for i, s in enumerate(sentences):
    print(f"  [{i}] {s}")
```

## Step 4–5: Load Model & Generate Embeddings

```python
# Step 4: Load embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Step 5: Generate embeddings
embeddings = model.encode(sentences)
print(f"Embedding shape: {embeddings.shape}  (10 sentences × 384 dimensions)")
```

## Step 6–8: Cosine Similarity Matrix & Heatmap

```python
# Step 6: Compute cosine similarity matrix
similarity_matrix = cosine_similarity(embeddings)

# Step 7: Print matrix
print("Cosine Similarity Matrix (rounded):\n")
print(np.round(similarity_matrix, 2))
```

```python
# Step 8: Plot heatmap WITH numbers
labels = [f"S{i}" for i in range(len(sentences))]

plt.figure(figsize=(10, 8))
ax = sns.heatmap(
    similarity_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    xticklabels=labels,
    yticklabels=labels,
    linewidths=0.5,
    vmin=0, vmax=1
)

plt.title("Cosine Similarity Matrix\n(S0–S2: Cricket | S3–S5: Cooking | S6–S9: Cybersecurity)", fontsize=13)
plt.xlabel("Sentences")
plt.ylabel("Sentences")
plt.tight_layout()
plt.savefig("similarity_heatmap.png", dpi=150, bbox_inches='tight')
plt.show()
print("\nHeatmap saved as similarity_heatmap.png")
```

## Step 9: Query — Find Most Similar Sentences

**Query:** *"The bowler took three wickets in one over"*

We expect the top matches to come from the **Cricket** topic cluster.


```python
# Step 9: Query sentence
query = "The bowler took three wickets in one over"

# Encode query
query_embedding = model.encode([query])

# Compute similarity with all sentences
scores = cosine_similarity(query_embedding, embeddings)[0]

# Get top 2 similar sentences (excluding itself at index 0)
similar_indices = scores.argsort()[-3:-1][::-1]

print("Query:", query)
print("\nTop 2 most similar sentences:")
for idx in similar_indices:
    print(f"  Sentence [{idx}]: {sentences[idx]}")
    print(f"  Similarity Score: {scores[idx]:.4f}\n")
```

## 📊 Observations & Conclusions

- **Intra-cluster similarity** (e.g., Cricket sentences S0–S2) is noticeably higher than **cross-cluster similarity**.
- The heatmap reveals clear **block structure** along the diagonal — sentences in the same topic group are more semantically similar.
- The query about a cricket bowler correctly retrieves other **cricket-related sentences** with the highest scores.
- `all-MiniLM-L6-v2` is a lightweight model that still captures semantic meaning well for short sentences.
